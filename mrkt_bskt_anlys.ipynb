```python
#converting data into list format.... in aprori
trans = []
for i in range(0,len(data)):
    trans.append([str(data.values[i,j]) for j in range(0,20) if str(data.values[i,j])!='0'])
```

```python
#Call apriori function which requires minimum support, confidance and lift,
# min length is combination of item default is 2".
rules = apriori(trans,min_support=0.003,min_confidance=0.2,min_lift=3,min_length=2)
```

```python
rules
```

```python
# all rules need to be converted in a list..
Results = list(rules)
Results
```

```python
data_results= pd.DataFrame(Results)
```

```python
data_results.head()
```

```python
support = data_results.support
```

```python
first = []
second= []
third = []
fourth = []
# loop number of rows time and append 1 by 1 value in a separate list..
# first and second element was frozenset which need to be converted in list..
for i in range(data_results.shape[0]):
    single = data_results['ordered_statistics'][i][0]
    first.append(list(single[0]))
    second.append(list(single[1]))
    third.append(single[2])
    fourth.append(single[3])
```

```python
#convert all four list into dataframe for further operation..
lhs = pd.DataFrame(first)
rhs= pd.DataFrame(second)
confidance=pd.DataFrame(third,columns=['Confidance'])
lift=pd.DataFrame(fourth,columns=['lift'])
```

```python
#concat all list together in a single dataframe
final = pd.concat([lhs,rhs,support,confidance,lift], axis=1)
final
```

```python
final.fillna(value=' ', inplace=True)
```

```python
#set column name
final.columns = ['lhs','lsh1','rhs',1,2,'support','confidance','lift']
final.head()
```

```python
#add all three column because those where the lhs itemset only
final['lhs'] = final['lhs']+str(", ")+final['lsh1']
final['rhs'] = final['rhs']+str(", ")+final[1]+str(", ")+final[2]
```
